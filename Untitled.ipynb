```python
# Missing value check
telecom_df.isna().sum()
```

```python
# To filter only the columns with missing values greater than 0
telecom_df.isna().sum()[telecom_df.isna().sum() > 0]
```

```python
# To get all the columns with numerical data
telecom_df.select_dtypes(include=['number']).columns
```

```python
# impute the missing values in the numerical columns with the mean of that column
# Loop through each numerical column and fill missing values with the mean of that column
for col in telecom_df.select_dtypes(include=['number']).columns:
    telecom_df[col] = telecom_df[col].fillna(telecom_df[col].mean())

```

```python
# Checking for cols with missing values
telecom_df.isna().sum()[telecom_df.isna().sum() > 0]
```
